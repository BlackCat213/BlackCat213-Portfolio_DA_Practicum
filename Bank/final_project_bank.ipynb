### Проверка на явные и неявные дубликаты.

```python
def check_duplicates(df):
    
    """this func checks ex-/implicit duplicates"""
    
    duplicates = df.duplicated()
    
    if duplicates.any():
        print('Дубликаты:')
        print(df[duplicates])
    else:
        print('Дубликаты не обнаружены')

 
    non_unique = df.columns[df.nunique() < len(df)]
    non_unique_duplicates = df.duplicated(subset=non_unique, keep=False)
    
    if non_unique_duplicates.any():
        print('Неявные дубликаты:')
        print(df[non_unique_duplicates])
    else:
        print('Неявные дубликаты не обнаружены')
```

```python
check_duplicates(data)
```

Проверка датасета на явные и неявные дубликаты не обнаружила таковых.


### Приведем все данные к одному разряду.

```python
def convert_to_32(df):
    
    """ this func converts 64 to 32 for saving memory and making all the same size."""
    
    int32 = [col for col in df.columns if col not in ['balance', 'est_salary', 'city', 'gender']]
    float32 = ['balance', 'est_salary']
    
    df[int32] = df[int32].astype(np.int32, errors='ignore')
    df[float32] = df[float32].astype(np.float32, errors='ignore')
    
    return df
```

```python
data = convert_to_32(data)
```

## Исследовательский анализ.

###  Распределение клиентов по городам.

```python
# создание сабплота 
fig, axes = plt.subplots(1, 2, figsize=(8, 4))

# данные по городам и количество клиентов в каждом городе
city_counts = data[data['churn'] == 0]['city'].value_counts()

# круговой график для процентного распределения клиентов по городам
axes[0].pie(city_counts, labels=city_counts.index, colors=colors, autopct='%1.1f%%')
axes[0].set_title('Неотточные клиенты по городам')

# данные по городам и  процент отточных клиентов в каждом городе
churn_city_counts = data[data['churn'] == 1]['city'].value_counts()

# круговой график для процентного распределения отточных клиентов по городам
axes[1].pie(churn_city_counts, labels=churn_city_counts.index, colors=colors, autopct='%1.1f%%')
axes[1].set_title('Распределение отточных клиентов по городам')

plt.tight_layout()
plt.show()
```

```python
city_churn_counts = data.groupby(['city', 'churn']).size().reset_index(name='count')

# таблица с абсолютными числами клиентов по полу и их статусам
city = pd.pivot_table(city_churn_counts, values='count', index='city', columns='churn', aggfunc=sum, fill_value=0)

city.columns = ['неотточные', 'отточные']
city['всего'] = city['неотточные'] + city['отточные']
```

```python
city['доля оттока'] = city['отточные'] / (city['отточные'] + city['неотточные'])
```

```python
rate_by_city = data.groupby('city')['churn'].mean() 
avg_churn = data['churn'].mean()


plt.figure(figsize=(7, 5))
rate_by_city.plot(kind='bar', color=colors, alpha=0.7, legend=False)
plt.axhline(avg_churn, color='red', linestyle='-', label='Средний отток по банку')
plt.xticks(rotation=45)
plt.xlabel('Город')
plt.ylabel('Доля оттока')
plt.title('Сравнение долей оттока по городам')
plt.legend(loc='upper right', bbox_to_anchor=(1.5, 1))
plt.show()
```

#### Расчет корреляции с использованием библиотеки phik.

Построим матрицу с использованием библиотеки phik, которая  позволяется вычислять взаимосвязи между различными типами признаков категориальными, непрерывными и их комбинациями.

```python
continious = ['age', 'score', 'balance', 'est_salary']
```

```python
corr_mx = data.drop('user_id', axis=1)
```

```python
phik_corr = corr_mx.phik_matrix(interval_cols=continious)
```

Данная тепловая карта указывает на слабую корреляцию между единицами собственности и скорингом, количеством продуктов и оценочным доходом и оттоком клиентов. По остальным параметрам зависимость либо очень слабая, либо отсутствует.

```python
churn_corr = phik_corr[['churn']].drop(['churn'])
churn_corr = churn_corr.sort_values(by='churn', ascending=False)

fig, ax = plt.subplots(figsize=(8, 6))
sns.heatmap(churn_corr, annot=True, cmap='coolwarm', center=0)
plt.title('Матрица корреляции(phik) по оттоку')
plt.xlabel('Отток')
plt.show()
```

Возраст таких клиентов может варьироваться от **27 до 47 лет**, а доход и баланс на счете обычно немного ниже, чем у тех клиентов, которые покидают банк. Вероятно, такие клиенты чувствуют удовлетворение от услуг банка и имеют лояльность к нему.

Второй типичный профиль - это отточный клиент. Чаще это **мужчина**.Они могут быть жителями **Ярославля или Ростова** и **менее склонны обладать кредитными картами**. Тем не менее, они также активно пользуются банковскими продуктами и **имеют несколько более высокий уровень собственности**, чем лояльные клиенты. Возраст отточных клиентов варьируется от **27 до 45 (с пиком в 32-33 года)** и от **50 до 60 лет**. Их **доход и баланс на счете обычно выше**, чем у клиентов, которые остаются в банке. Вероятно, эти клиенты находят более выгодные предложения или условия в других банках и принимают решение перейти к конкурентам.

Отмечается, что между различными характеристиками клиентов присутствует слабая или очень слабая корреляция.

## Статистический анализ.

В данном разделе мы проверим гипотезы о:
- 1. различии дохода между уходящими клиентами и оставшимися;
- 2. различие долей полов клиентов, которые уходят и теми, кто остается;
- 3. различии  оттока между городами;
- 4. различии возрастов между оставшимися клиентами и уходящими.

Подготовим данные для тестов, разделим клиентов на уходящих и оставшихся.

```python
churned = data[data['churn'] == 1]
retained = data[data['churn'] == 0]
```

### Гипотеза о различаи дохода клиентов.

Сформулируем гипотезу:
   - H0: Нет значимых различий в доходе между клиентами, которые ушли, и клиентами, которые остались.
   - H1: Существуют значимые различия в доходе между клиентами, которые ушли, и клиентами, которые остались.

Наиболее частые случаи применения t-критерия связаны с проверкой равенства средних значений в двух выборках. Выбор данного теста так же предполагает, что у нас нормальное распределение и отсутствие существенного числа выбросов.

```python
t_statistic, p_value = ttest_ind(churned['est_salary'], retained['est_salary'], alternative='two-sided')

alpha = 0.05

if p_value < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не удалось отвергнуть нулевую гипотезу')
    
print('p-значение:', p_value)
print('t-критерий:',t_statistic)
```

Если же данные не подчиняются нормальному распределению, то стоит использовать непараметрические тесты, которые основаны на рангах данных и не требуют предположения о форме распределения. В нашем случае мы проведем тест Манна-Уитни. 

```python
u_statistic, p_value = mannwhitneyu(churned['est_salary'], retained['est_salary'], alternative='two-sided')

alpha = 0.05

if p_value < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не удалось отвергнуть нулевую гипотезу')
    
print('p-значение:', p_value)
print('u-критерий:',u_statistic)
```

Определим, почему есть разница между тестами. Так как, проверка с использованием t-критерия предполагает, что у нас нормально распределение, проверим его нормальность с помощью теста Шапиро-Уилка.

```python
stat, p_value = shapiro(data['est_salary'])

alpha = 0.05

if p_value > alpha:
    print('Данные подчиняются нормальному распределению')
else:
    print('Данные не подчиняются нормальному распределению')
    
print(f'Статистика теста: {stat}')
print(f'p-значение: {p_value}')
```

QQ-график подтверждает, что данные о доходе не подчиняются нормальному распределению, а следовательно за основу возьмем результаты непараметрического теста Манна-Уитни, который отвергает нулевую гипотезу, о том, что нет статистически значимых различий между доходами оставшимися и уходящими клиентами.

**Вывод:**

### Гипотеза о различие долей полов клиентов, которые уходят и теми, кто остается.

Сформулируем гипотезу:
   - H0: Нет статистических различий долей полов клиентов, которые уходят и теми, кто остается.
   - H1: Существуют статистические различия долей полов клиентов, которые уходят и теми, кто остается.

Перед тем, как приступить к проверке данной гипотезу проведем тест на независимость наблюдений хи-квадрат для оценки связи между категориальными переменными.

```python
males = data[data['gender'] == 'М']
females = data[data['gender'] == 'Ж']
```

```python
# наблюдаемое кол-во
observed_counts = pd.DataFrame({'male': [len(males)], 'female': [len(females)]})

# ожидаемое кол-во
expected_counts = observed_counts.sum() / 2
alpha = 0.05
# тест на хи-квадрат
chi2, p_value, _, _ = chi2_contingency(observed_counts)

if p_value < alpha:
    print('Отвергаем гипотезу о случайности выборки, есть стат. значимые доказательства наличия неравномерности в данных')
else:
    print('Не удалось отвергнуть гипотезу о случайности выборки, нет достаточных стат. доказательств наличия неравномерности')
    

print('\nНаблюдаемое количество:')
print(observed_counts)
print('\nОжидаемое количество:')
print(expected_counts)
print(f'\nЗначение хи-квадрат: {chi2}')
print(f'p-значение: {p_value}')
```

Нам не удалось опровергнуть случайность выборки, следовательно, мы можем приступить к дальнейшему выполнению Z-тест, один из критериев которого, наблюдения в каждой группе независимы друг от друга, то есть, каждое наблюдение является случайной выборкой из ген.совокупности и отсутствует связь или влияние между ними.

```python
# Вычисление доли успешных событий в каждой группе
p1 = males['churn'].mean()
p2 = females['churn'].mean()

alpha = 0.05

# Вычисление общей пропорции успешных событий в обоих группах
p_combined = (males['churn'].sum() + females['churn'].sum()) / (len(males) + len(females))

# Вычисление разницы пропорций
difference = p1 - p2

# Вычисление стандартного отклонения разницы пропорций
std_dev = (p_combined * (1 - p_combined) * (1 / len(males) + 1 / len(females))) ** 0.5

# Вычисление статистики z
z_value = difference / std_dev

# Вычисление двустороннего p-значения
p_value = 2 * (1 - stats.norm.cdf(abs(z_value)))

if p_value < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не удалось отвергнуть нулевую гипотезу')
    
print('p-значение:', p_value)
print('z-критерий:',z_value)
```

**Вывод:** 
- мы не смогли опровергнуть гипотезу о наличаи зависимости наблюдений;
- нам удалось отвергнуть нулевую гипотезу о том, что нет статистических различий долей полов клиентов, которые уходят и теми, кто остается.

####  Гипотеза о различие долей  клиентов мужского пола, которые уходят и теми, кто остается.

Сформулируем гипотезу:
   - H0: Нет статистических различий долей клиентов мужского пола, которые уходят и теми, кто остается.
   - H1: Существуют статистические различия долей клиентов мужского пола, которые уходят и теми, кто остается.

```python
males_churn = males[males['churn'] == 1]
males_not_churn = males[males['churn'] == 0]

p1 = males_churn.shape[0] / males.shape[0]
p2 = males_not_churn.shape[0] / males.shape[0]

alpha = 0.05

# Вычисление разницы пропорций
difference = p1 - p2

# Вычисление стандартного отклонения разницы пропорций
std_dev = (p1 * (1 - p1) / males.shape[0] + p2 * (1 - p2) / males.shape[0]) ** 0.5

# Вычисление статистики z
z_value = difference / std_dev
# скор. уровень значимости после поправки Шидака
corrected_alpha = 1 - (1 - alpha)**(1/2)  # число сравнений

# Вычисление двустороннего p-значения
p_value = 2 * (1 - stats.norm.cdf(abs(z_value)))

if p_value < corrected_alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не удалось отвергнуть нулевую гипотезу')
    
print('p-значение для мужчин:', p_value)
print('z-критерий для мужчин:', z_value)
```

**Вывод:** мы отвергли гипотезу об отсутствии значимых статистических различий долей клиентов мужского пола, которые уходят и теми, кто остается.

#### Гипотеза о различие долей  клиентов женского пола, которые уходят и теми, кто остается.

Сформулируем гипотезу:

- H0: Нет статистических различий долей клиентов женского пола, которые уходят и теми, кто остается.
- H1: Существуют статистические различия долей клиентов женского пола, которые уходят и теми, кто остается.

###  Гипотеза о различии  оттока между городами.</span>

Сформулируем гипотезу:

- H0: Нет статистически значимых различий в оттоке между городами.
- H1: Существуют статистически значимые различия в оттоке между городами.

 Для проверки гипотезы проведем непараметрический тест, они обычно менее мощные, но могут быть более универсальными в случае нарушения предположений о данных. В нашем случае выберем тест Краскела-Уоллиса. Данный тест может использоваться как с независимыми, так и с зависимыми выборками (например, с повторными измерениями или парными наблюдениями).

```python
alpha=0.05

ryb = data[data['city'] == 'Рыбинск']
yar = data[data['city'] == 'Ярославль']
ros = data[data['city'] == 'Ростов']

# тест Краскела-Уоллиса
statistic, p_value = stats.kruskal(ryb['churn'], yar['churn'], ros['churn'])


if p_value < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не удалось отвергнуть нулевую гипотезу') 
    
    
print("Статистика теста: ", statistic)
print("p-значение: ", p_value)
```

```python
city_data = {}
for city in data["city"].unique():
    city_data[city] = data[data["city"] == city]["score"]


results = {}
alpha_shidaka = 1 - (1 - alpha) ** (1/6)

for city1 in city_data.keys():
    for city2 in city_data.keys():
        if city1 != city2:
            z_stat, p_val = zscore(city_data[city1]), zscore(city_data[city2])
            is_sig = p_val < alpha
            results[(city1, city2)] = (z_stat, p_val, is_sig)
            if p_value < alpha:
                print('Отвергаем нулевую гипотезу')
            else:
                print('Не удалось отвергнуть нулевую гипотезу') 


for cities, (z_stat, p_val, is_sig) in results.items():
    print(f"Города: {cities}")
    print(f"Z-статистика: {z_stat}")
    print(f"p-значение: {p_val}")
    print(f"Статистически значимо: {is_sig}")
    print("=============================")

print(f"Скорректированная альфа Шидака: {alpha_shidaka}")
```

```python
conditions = [
    'equity >= 3',
    'last_activity == 1',
    'gender == "М"',
    'est_salary.between(100000, 250000)',
    'credit_card == 0',
    'products>=2',
    'age.between(25, 35)',
    'score.between(820, 920)',
    'balance > 900000'
    ]
```

```python
def calculate_churn(df, conditions, min_count, max_count, chrn): 
    
        
    """this func creates all possible combs(max conditions = 3 per 1 check),
    checks all through and calculate chrn rate and choose higher
    and gives info abt combs
    """
    
    results = []
    
    # Все возможные комбинации условий, max 3 условия за комбинацию
    all_combinations = []
    for r in range(1, min(len(conditions)+1, 4)):
        combinations = list(itertools.combinations(conditions, r))
        all_combinations.extend(combinations)
    
    # Все комбинации условий и расчет оттока
    for combination in all_combinations:
        query = ' and '.join(combination)
        result = df.query(query).agg({'churn': ['mean', 'count']})

        # Проверка значения churn:mean и count
        churn_mean = result['churn']['mean']
        count = result['churn']['count']
        if churn_mean > chrn and count >= min_count and count <= max_count:
            # Отбираем первые слова из каждой комбинации
            first_words = [word.split()[0] for word in combination]
            combination_info = ', '.join(first_words)
            result['combination_info'] = combination_info
            results.append(result)

    return results
```
